# Estado del Arte — Clasificación del Bosón de Higgs

Revisión de trabajos que han abordado el problema de clasificación binaria señal/fondo usando el dataset HIGGS de UCI (Baldi et al., 2014) o que han propuesto algoritmos de ML validados en este dataset como benchmark.

La búsqueda se realizó en las bases de datos Elsevier (ScienceDirect), IEEE Xplore y ACM Digital Library, complementada con Semantic Scholar y Google Scholar usando las consultas: *"HIGGS dataset machine learning classification"*, *"Higgs boson detection deep learning"*, y *"gradient boosting HIGGS benchmark"*.

## Paper 1: Searching for Exotic Particles in High-Energy Physics with Deep Learning

| Campo | Detalle |
|-------|----------|
| **Autores** | Baldi, P., Sadowski, P., Whiteson, D. |
| **Revista / Año** | Nature Communications, 2014 |
| **Paradigma** | Supervisado — Clasificación binaria |
| **Modelos usados** | Deep NN (5 capas con dropout), Shallow NN, Regresión Logística, Boosted Decision Trees (TMVA) |
| **Metodología de validación** | Hold-out (entrenamiento en ~10.5M muestras, test en 500K) |
| **Métricas** | AUC-ROC, Accuracy, significancia de descubrimiento |
| **Resultados** | Deep NN: AUC = 0.885; Shallow NN: AUC = 0.816; Regresión Logística: AUC = 0.733 |

**Resumen:** Trabajo fundacional que introduce el dataset HIGGS como banco de prueba para comparar clasificadores supervisados en física de altas energías. Los autores demuestran que redes neuronales profundas (5 capas, dropout) superan significativamente a métodos clásicos y a redes poco profundas. Un hallazgo clave es que las redes profundas entrenadas solo con features de bajo nivel (kinemáticas) alcanzan un AUC ≈ 0.880, muy cercano al obtenido con el conjunto completo (0.885), lo que sugiere que las redes profundas pueden aprender automáticamente las transformaciones no lineales que los físicos codifican manualmente en las features de alto nivel. Este trabajo establece la línea base de referencia del campo y proporciona el protocolo experimental que inspira nuestro proyecto.

## Paper 2: Higgs Boson Discovery using Machine Learning Methods with PySpark

| Campo | Detalle |
|-------|----------|
| **Autores** | Azhari, M., Abarda, A., Ettaki, B., Zerouaoui, J., Dakkon, M. |
| **Revista / Año** | Procedia Computer Science, 2020 |
| **Paradigma** | Supervisado — Clasificación binaria |
| **Modelos usados** | Regresión Logística, Árbol de Decisión, Random Forest, Gradient Boosted Tree |
| **Metodología de validación** | Hold-out (80/20 split), preprocesamiento con Spark ML |
| **Métricas** | AUC-ROC, Accuracy |
| **Resultados** | Mejor modelo: Gradient Boosted Tree con AUC ≈ 0.705, Accuracy ≈ 0.706 |

**Resumen:** Este trabajo aborda el mismo problema de clasificación del dataset HIGGS pero desde una perspectiva de procesamiento distribuido. Los autores implementan cuatro modelos clásicos de ML en PySpark (MLlib) para manejar la escala de datos del dataset completo. Sus resultados muestran que el Gradient Boosted Tree supera a Random Forest, Árbol de Decisión y Regresión Logística en este entorno. La diferencia con Baldi et al. (AUC 0.885 vs 0.705) ilustra cómo el protocolo experimental, el preprocesamiento distribuido y los recursos de entrenamiento influyen significativamente en el resultado final, más allá de la elección del modelo.

## Paper 3: XGBoost: A Scalable Tree Boosting System

| Campo | Detalle |
|-------|----------|
| **Autores** | Chen, T., Guestrin, C. |
| **Revista / Año** | KDD (ACM SIGKDD), 2016 |
| **Paradigma** | Supervisado — Clasificación y regresión |
| **Modelos usados** | XGBoost (Extreme Gradient Boosting) |
| **Metodología de validación** | Multiple datasets, HIGGS entre ellos (validación estándar) |
| **Métricas** | AUC-ROC, Accuracy, tiempo de ejecución |
| **Resultados** | AUC ≈ 0.882 en HIGGS (con el dataset completo de 11M), superando a los métodos previos |

**Resumen:** Chen y Guestrin presentan XGBoost, un sistema de gradient boosting optimizado que introduce regularización explícita, paralelismo a nivel de features y manejo eficiente de valores faltantes. El dataset HIGGS es uno de sus benchmarks principales, donde demuestran que XGBoost iguala o supera a las redes profundas de Baldi et al. con menor costo computacional. Este trabajo es particularmente relevante para nuestro proyecto porque uno de los modelos evaluados es XGBoost, y establece la expectativa de que los métodos de gradient boosting modernos pueden ser competitivos frente a deep learning en datasets tabulares de física de partículas.

## Paper 4: LightGBM: A Highly Efficient Gradient Boosting Decision Tree

| Campo | Detalle |
|-------|----------|
| **Autores** | Ke, G., Meng, Q., Finley, T., Wang, T., Chen, W., Ma, W., Ye, Q., Liu, T.-Y. |
| **Revista / Año** | NeurIPS, 2017 |
| **Paradigma** | Supervisado — Clasificación y regresión |
| **Modelos usados** | LightGBM (Gradient Boosting leaf-wise) |
| **Metodología de validación** | Múltiples datasets incluyendo HIGGS, comparación con XGBoost y otros |
| **Métricas** | AUC-ROC, Accuracy, velocidad de entrenamiento |
| **Resultados** | AUC ≈ 0.881 en HIGGS, con velocidad de entrenamiento ~20× más rápida que XGBoost |

**Resumen:** Ke et al. introducen LightGBM, que se diferencia de XGBoost en dos aspectos clave: (1) crecimiento de árboles por hoja (leaf-wise) en lugar de por nivel (level-wise), produciendo árboles más profundos y precisos, y (2) técnicas de muestreo basadas en gradientes (Gradient-based One-Side Sampling) y agrupación exclusiva de features (Exclusive Feature Bundling) para acelerar el entrenamiento. LightGBM alcanza un AUC ≈ 0.881 en el dataset HIGGS con un costo computacional significativamente menor. Este trabajo respalda la inclusión de LightGBM en nuestro conjunto de modelos evaluados y contextualiza los resultados obtenidos.

## Métricas No Vistas en Clase

### AUC-ROC (Área Bajo la Curva ROC)

La curva ROC (*Receiver Operating Characteristic*) grafica la Tasa de Verdaderos Positivos (TPR) vs la Tasa de Falsos Positivos (FPR) para distintos umbrales de clasificación:

$$\text{TPR} = \frac{TP}{TP + FN}, \quad \text{FPR} = \frac{FP}{FP + TN}$$

El AUC-ROC es el área bajo esta curva y se interpreta como la probabilidad de que el modelo asigne un score mayor a una muestra positiva que a una negativa. Valores: 0.5 = aleatorio, 1.0 = perfecto.

Esta métrica es estándar en la literatura del dataset HIGGS y es robusta ante el desbalance de clases. En nuestro proyecto es la métrica principal de comparación, complementada con Precision, Recall, F1-Score y Accuracy como métricas secundarias.
